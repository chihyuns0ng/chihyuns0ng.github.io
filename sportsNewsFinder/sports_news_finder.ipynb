```python
def news_finder(Y, M, D):
    
    day = date(Y, M, D).strftime("%Y%m%d")
    url = f"https://sports.news.naver.com/ranking/list.nhn?type=popular&date={day}"
    
    response = requests.get(url, headers={"user-agent" : "Mozilla/5.0"})
    sports_json = response.json()
    sports_list = sports_json["rankingNewsList"]["list"]
    
    df = pd.DataFrame(sports_list)
    #문자열 표기 길이 조정
    # pd.options.display.max_colwidth = 100
    
    df = df.set_index("rank")
    df = df.drop(["date", "device", "oid", "aid", "type", "cnt", "newsDateTime", "thumbnail", "modifiedDateTime", "subContent"], axis=1)
    df.rename(columns={"totalCount" : "조회수"}, inplace=True)
    
    file_name = f"sportNews_{day}.csv"
    df.to_csv(file_name)
    
    print(f"{Y}년 {M}월 {D}일의 뉴스 랭킹입니다." )
    return pd.read_csv(file_name)
```
